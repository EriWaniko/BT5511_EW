```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from difflib import SequenceMatcher

# -----------------------------
# 1. Set data folder
# -----------------------------

DATA_DIR = Path(r"C:\Users\eriat\OneDrive\School\MSc\University of Aberdeen\Term 2 (Jan-Mar)\Advanced Biotechnology with Computation\Assignment\Biocomputation Report")

files = sorted(DATA_DIR.glob("*_data_file.txt"))

if len(files) == 0:
    raise FileNotFoundError("No *_data_file.txt files found. Check the DATA_DIR path.")

print(f"Number of files found: {len(files)}")
for file in files:
    print(file.name)

# -----------------------------
# 2. Read and combine files
# -----------------------------

raw = pd.concat(
    [pd.read_csv(f, sep="\t").assign(Source_File=f.name) for f in files],
    ignore_index=True
)

print(raw.shape)
display(raw.head())

# -----------------------------
# 3. Define mutation functions
# -----------------------------

def validate_dna(seq):
    """Check that a DNA sequence contains only A, T, C and G."""
    seq = str(seq).upper()
    if not re.fullmatch(r"[ATCG]+", seq):
        raise ValueError(f"Invalid DNA sequence found: {seq[:50]}")
    return seq


def classify_mutation(wt, mut):
    """Classify mutation as substitution, insertion or deletion."""
    if len(wt) == len(mut):
        return "Substitution" if wt != mut else "No difference"
    elif len(mut) > len(wt):
        return "Insertion"
    else:
        return "Deletion"


def first_mutation_position(wt, mut):
    """Find the first mutation position using 1-based indexing."""
    matcher = SequenceMatcher(None, wt, mut)
    positions = []

    for tag, i1, i2, j1, j2 in matcher.get_opcodes():
        if tag != "equal":
            positions.append(i1 + 1)

    return min(positions) if positions else None


def mutation_size(wt, mut):
    """Measure mutation size."""
    if len(wt) != len(mut):
        return abs(len(mut) - len(wt))
    return sum(base1 != base2 for base1, base2 in zip(wt, mut))


def mutation_region(first_pos):
    """Classify mutation as promoter or coding region."""
    if first_pos is None:
        return "No mutation"
    return "Promoter" if first_pos <= 1000 else "Coding"

# -----------------------------
# 4. Validate and classify sequences
# -----------------------------

raw["WildType.Sequence"] = raw["WildType.Sequence"].apply(validate_dna)
raw["Mutant.Sequence"] = raw["Mutant.Sequence"].apply(validate_dna)

raw["WT.Length"] = raw["WildType.Sequence"].str.len()
raw["Mutant.Length"] = raw["Mutant.Sequence"].str.len()

raw["Mutation.Class"] = raw.apply(
    lambda row: classify_mutation(row["WildType.Sequence"], row["Mutant.Sequence"]),
    axis=1
)

raw["First.Mutation.Position"] = raw.apply(
    lambda row: first_mutation_position(row["WildType.Sequence"], row["Mutant.Sequence"]),
    axis=1
)

raw["Mutation.Size"] = raw.apply(
    lambda row: mutation_size(row["WildType.Sequence"], row["Mutant.Sequence"]),
    axis=1
)

raw["Mutation.Region"] = raw["First.Mutation.Position"].apply(mutation_region)

# -----------------------------
# 5. Calculate mean expression and viability changes
# -----------------------------

mRNA_wt = ["mRNA.Expression.WT.Rep1", "mRNA.Expression.WT.Rep2", "mRNA.Expression.WT.Rep3"]
mRNA_mut = ["mRNA.Expression.Mut.Rep1", "mRNA.Expression.Mut.Rep2", "mRNA.Expression.Mut.Rep3"]

protein_wt = ["Protein.Expression.WT.Rep1", "Protein.Expression.WT.Rep2", "Protein.Expression.WT.Rep3"]
protein_mut = ["Protein.Expression.Mut.Rep1", "Protein.Expression.Mut.Rep2", "Protein.Expression.Mut.Rep3"]

viability_wt = ["CellViability.WT.Rep1", "CellViability.WT.Rep2", "CellViability.WT.Rep3"]
viability_mut = ["CellViability.Mut.Rep1", "CellViability.Mut.Rep2", "CellViability.Mut.Rep3"]

raw["mRNA.WT.Mean"] = raw[mRNA_wt].mean(axis=1)
raw["mRNA.Mut.Mean"] = raw[mRNA_mut].mean(axis=1)
raw["mRNA.Change"] = raw["mRNA.Mut.Mean"] - raw["mRNA.WT.Mean"]

raw["Protein.WT.Mean"] = raw[protein_wt].mean(axis=1)
raw["Protein.Mut.Mean"] = raw[protein_mut].mean(axis=1)
raw["Protein.Change"] = raw["Protein.Mut.Mean"] - raw["Protein.WT.Mean"]

raw["Viability.WT.Mean"] = raw[viability_wt].mean(axis=1)
raw["Viability.Mut.Mean"] = raw[viability_mut].mean(axis=1)
raw["Viability.Change"] = raw["Viability.Mut.Mean"] - raw["Viability.WT.Mean"]

# -----------------------------
# 6. Create priority score
# -----------------------------

raw["Priority.Score"] = (
    raw["Viability.Change"].abs().rank(pct=True) * 0.50
    + raw["mRNA.Change"].abs().rank(pct=True) * 0.20
    + raw["Protein.Change"].abs().rank(pct=True) * 0.20
    + raw["Mutation.Size"].rank(pct=True) * 0.10
)

# -----------------------------
# 7. Create final summary table
# -----------------------------

summary_cols = [
    "Gene",
    "Source_File",
    "Mutation.Class",
    "Mutation.Region",
    "First.Mutation.Position",
    "Mutation.Size",
    "mRNA.Change",
    "Protein.Change",
    "Viability.Change",
    "Priority.Score"
]

summary = raw[summary_cols].sort_values("Priority.Score", ascending=False)

display(summary)

summary.to_csv(DATA_DIR / "integrated_mutation_screening_summary.csv", index=False)

# -----------------------------
# 8. Visualise results
# -----------------------------

# Figure 1: viability change by mutation class
fig, ax = plt.subplots(figsize=(8, 5))

classes = sorted(summary["Mutation.Class"].unique())
data = [summary.loc[summary["Mutation.Class"] == cls, "Viability.Change"] for cls in classes]

ax.boxplot(data, labels=classes)
ax.set_title("Cell viability change by mutation class")
ax.set_xlabel("Mutation class")
ax.set_ylabel("Mutant mean viability - wild-type mean viability")

plt.tight_layout()
plt.show()

# Figure 2: mRNA change versus protein change
fig, ax = plt.subplots(figsize=(8, 5))

for cls in classes:
    subset = summary[summary["Mutation.Class"] == cls]
    ax.scatter(subset["mRNA.Change"], subset["Protein.Change"], label=cls)

ax.axhline(0, linewidth=1)
ax.axvline(0, linewidth=1)
ax.set_title("mRNA change versus protein expression change")
ax.set_xlabel("Mutant mean mRNA - wild-type mean mRNA")
ax.set_ylabel("Mutant mean protein - wild-type mean protein")
ax.legend(title="Mutation class")

plt.tight_layout()
plt.show()

# Figure 3: viability change distribution
fig, ax = plt.subplots(figsize=(8, 5))

ax.hist(summary["Viability.Change"], bins=10)
ax.set_title("Distribution of cell viability changes")
ax.set_xlabel("Mutant mean viability - wild-type mean viability")
ax.set_ylabel("Number of genes")

plt.tight_layout()
plt.show()
```
